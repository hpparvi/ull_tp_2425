**GIF FOR 3 BODY PROBLEM**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Cargar datos desde el archivo
data = np.loadtxt('/content/orbits_3_particles.txt')

# Separar columnas
time = data[:, 0]         # Tiempo
particle_id = data[:, 1]  # Identificador de la partícula
x = data[:, 2]            # Coordenada x
y = data[:, 3]            # Coordenada y

# Crear listas para almacenar las trayectorias de cada partícula
x1, y1 = [], []
x2, y2 = [], []
x3, y3 = [], []

# Organizar las coordenadas según el ID de la partícula
for i in range(len(particle_id)):
    if particle_id[i] == 1:
        x1.append(x[i])
        y1.append(y[i])
    elif particle_id[i] == 2:
        x2.append(x[i])
        y2.append(y[i])
    elif particle_id[i] == 3:
        x3.append(x[i])
        y3.append(y[i])


# Crear la figura y los ejes para la animación
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.2, 1.2)

# Configuración de la gráfica: se define la línea del rastro y el marcador de cada partícula
trail1, = ax.plot([], [], '-', color='blue', alpha=0.5, label='Particle 1')
trail2, = ax.plot([], [], '-', color='red', alpha=0.5, label='Particle 2')
trail3, = ax.plot([], [], '-', color='green', alpha=0.5, label='Particle 3')
particle1, = ax.plot([], [], 'o', color='blue')  # marcador actual de partícula 1
particle2, = ax.plot([], [], 'o', color='red')   # marcador actual de partícula 2
particle3, = ax.plot([], [], 'o', color='green')   # marcador actual de partícula 3

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
ax.grid()

# Función de inicialización para la animación
def init():
    trail1.set_data([], [])
    trail2.set_data([], [])
    trail3.set_data([], [])
    particle1.set_data([], [])
    particle2.set_data([], [])
    particle3.set_data([], [])
    return trail1, trail2, trail3, particle1, particle2, particle3

# Función para actualizar los datos en cada fotograma
def update(frame):
    # Actualizar el rastro de las partículas hasta el fotograma actual
    trail1.set_data(x1[:frame], y1[:frame])
    trail2.set_data(x2[:frame], y2[:frame])
    trail3.set_data(x3[:frame], y3[:frame])

    # Actualizar la posición actual de las partículas
    particle1.set_data(x1[frame-1], y1[frame-1])
    particle2.set_data(x2[frame-1], y2[frame-1])
    particle3.set_data(x3[frame-1], y3[frame-1])

    return trail1, trail2, trail3, particle1, particle2, particle3

# Crear la animación
ani = FuncAnimation(fig, update, frames=np.arange(0, len(x1), 1), init_func=init, blit=True)

# Guardar la animación como un GIF
ani.save('/content/orbit_animation_3_60fps.gif', writer='imagemagick', fps=60)

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Cargar datos desde el archivo
data = np.loadtxt('/content/orbits_2_particles.txt')

# Separar columnas
time = data[:, 0]         # Tiempo
particle_id = data[:, 1]  # Identificador de la partícula
x = data[:, 2]            # Coordenada x
y = data[:, 3]            # Coordenada y

# Crear listas para almacenar las trayectorias de cada partícula
x1, y1 = [], []
x2, y2 = [], []

# Organizar las coordenadas según el ID de la partícula
for i in range(len(particle_id)):
    if particle_id[i] == 1:
        x1.append(x[i])
        y1.append(y[i])
    elif particle_id[i] == 2:
        x2.append(x[i])
        y2.append(y[i])

# Crear la figura y los ejes para la animación
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(min(min(x1), min(x2)) - 1, max(max(x1), max(x2)) + 1)
ax.set_ylim(min(min(y1), min(y2)) - 1, max(max(y1), max(y2)) + 1)

# Configuración de la gráfica: se define la línea del rastro y el marcador de cada partícula
trail1, = ax.plot([], [], '-', color='blue', alpha=0.5, label='Particle 1')
trail2, = ax.plot([], [], '-', color='red', alpha=0.5, label='Particle 2')
particle1, = ax.plot([], [], 'o', color='blue')  # marcador actual de partícula 1
particle2, = ax.plot([], [], 'o', color='red')   # marcador actual de partícula 2

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
ax.grid()

# Función de inicialización para la animación
def init():
    trail1.set_data([], [])
    trail2.set_data([], [])
    particle1.set_data([], [])
    particle2.set_data([], [])
    return trail1, trail2, particle1, particle2

# Función para actualizar los datos en cada fotograma
def update(frame):
    # Actualizar el rastro de las partículas hasta el fotograma actual
    trail1.set_data(x1[:frame], y1[:frame])
    trail2.set_data(x2[:frame], y2[:frame])

    # Actualizar la posición actual de las partículas
    particle1.set_data(x1[frame-1], y1[frame-1])
    particle2.set_data(x2[frame-1], y2[frame-1])

    return trail1, trail2, particle1, particle2

# Crear la animación
ani = FuncAnimation(fig, update, frames=np.arange(0, len(x1), 1), init_func=init, blit=True)

# Guardar la animación como un GIF
ani.save('/content/orbit_animation_2_60fps.gif', writer='imagemagick', fps=60)

plt.show()
```
