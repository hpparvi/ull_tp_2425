## Stable Circular Orbit 

### 2 particles

```python
filename = 'test_2part_stable_orbit_positions'
pos, ids, mass = utils.read_data(filename)
```

```python
colors = utils.generate_colors(len(ids), cmap='gnuplot_r')
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
xyz = [(0, 1), (0, 2), (1, 2)]
for i, particle in enumerate(ids):
    for j in range(3):
        if pos[:, 0][particle][0] - pos[:, 0][particle][-1] < 0.001:
            ax[j].plot(pos[:, xyz[j][0]][ids[i]], pos[:, xyz[j][1]][ids[i]], 'o', color=colors[i], ms=20)
        else:
            ax[j].plot(pos[:, xyz[j][0]][ids[i]], pos[:, xyz[j][1]][ids[i]], '-', color=colors[i])
            ax[j].plot(pos[:, xyz[j][0]][ids[i]][-1], pos[:, xyz[j][1]][ids[i]][-1], 'o', color=colors[i], ms=10)

fig, ax = utils.labels_plots(fig, ax)
fig.savefig('../output/figures/' + filename + '.png')
```

### 5 particles

```python
filename = 'test_5part_stable_orbit_positions'
pos, ids, mass = utils.read_data(filename)
```

```python
colors = utils.generate_colors(len(ids), cmap='gnuplot_r')
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

for i, particle in enumerate(ids):
    for j in range(3):
        if np.linalg.norm(pos[particle][0] - pos[particle][-1]) < 1e-3:
            ax[j].plot(pos[:, xyz[j][0]][ids[i]], pos[:, xyz[j][1]][ids[i]], 'o', color=colors[i], ms=20)
        else:
            ax[j].plot(pos[:, xyz[j][0]][ids[i]], pos[:, xyz[j][1]][ids[i]], '-', color=colors[i])
            ax[j].plot(pos[:, xyz[j][0]][ids[i]][-1], pos[:, xyz[j][1]][ids[i]][-1], 'o', color=colors[i], ms=10)
        ax[j].set_xlim(-2.2, 2.2)
        ax[j].set_ylim(-2.2, 2.2)

fig, ax = utils.labels_plots(fig, ax)

fig.savefig('../output/figures/' + filename + '.png')
```
